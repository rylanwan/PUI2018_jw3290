```python
test = '''
SELECT start_station_name, COUNT(cartodb_id) as trips,
(SQRT(POW(start_station_latitude -40.7577,2) + POW(start_station_longitude +73.9857,2))*111000) as distance
FROM htv210.citibike_feb_2015
WHERE (SQRT(POWER(start_station_latitude -40.7577,2) + POWER(start_station_longitude +73.9857,2))*111000) < 500
GROUP BY start_station_name, start_station_latitude, start_station_longitude
ORDER BY distance
'''

table = get_data(test)
table
```

**Task 4 â€” Putting it all together**
